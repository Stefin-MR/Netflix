### Data Analysis and Visualization

```python
# Create a DataFrame that has the greatest time point for each mouse

grouped_DF = nodupe_df.groupby(['Mouse ID'])
GTP = grouped_DF["Timepoint"].max()
GTP_df = pd.DataFrame(GTP)
#  Reset the index of the DataFrame
nodupe_df.reset_index()
# Display the DataFrame 
nodupe_df
```

```python
# JMerge the newly created `.max()` DataFrame with the cleaned DataFrame.
New_clean_df=pd.merge(nodupe_df,GTP_df,on=("Mouse ID","Timepoint"),how="right")
New_clean_df
#Timepoint_y = max
```
