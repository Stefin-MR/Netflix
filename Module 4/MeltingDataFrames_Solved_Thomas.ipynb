```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Get the file path for running team one
team1_file = os.path.join('../','Resources','team-1.csv')
```

```python
# Read the CSV file into a DataFrame
team1_df= pd.read_csv(team1_file)
team1_df
```

```python
# Plot the DataFrame before the uisng melt()
team1_df.plot.bar()
```

```python
# Melt the DataFrame
team1_df.melt()
```

```python
# Melt the DataFrame using the runner name.
team1_df.melt('runner_name')
```

```python
# Melt the DataFrame using the coach name.
team1_df.melt('coach_name')
```

```python
# Remove coach name from original DataFrame
team3_df=team1_df.drop(columns='coach_name')
team3_df
```

```python
# Melt the updated original DataFrame using the runner_name
# column, and add a var_name parameter and a value_name 
# parameter to rename the columns.
team3_df=team1_df.melt('runner_name')
team3_df
team3_df=team3_df.rename(columns={'variable':'var_name','value':'value_name'})
team3_df=team3_df.drop(team3_df.index[[0,1,2,3]])
team3_df
```

```python
# Plot a bar chart to show how the data has changed after melting it.
team3_df.plot.bar()
```

```python
# Save melted DataFreame to a CSV file
team3_df.to_csv('MeltingDF.csv',sep = '|')
```

```python
# Convert the melted DataFrame from long form back to wide form using 
# the pivot_table() function.
pd.pivot_table(team3_df,index='runner_name',columns='var_name')
```
