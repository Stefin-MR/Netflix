**Question 2: Create a code block that determines the mean, median, and mode of the final grades from the `student.json` file. (5 points)**


````python
# Answer 2
import pandas as pd
import numpy as np
import scipy.stats as sts

students_grades = pd.read_json('Resources/student.json')
print('The mean is ', students_grades['Final Grade'].mean())
print('The median is ', students_grades['Final Grade'].median())
print('The mode is ', students_grades['Final Grade'].mode())
````

**Question 3: Use Plotly to create a histogram of the final grades from the `student.json` file. (5 points)**


````python
# Answer 3
import pandas as pd
import plotly.express as px

students_grades = pd.read_json('Resources/student.json')
fig = px.histogram(students_grades["Final Grade"])
fig.update_layout(xaxis_title="Final Grade", yaxis_title="Counts")
fig.show()
````

**Question 5: Use Plotly to create a code block that performs the following operations: (10 points)**

   1. Import the `G20_crop_production_2018-2020.csv` file into a Pandas DataFrame.
   
   2. Create three box plots showing the `Value(tonnes of HA)` by `Crop` on the same chart where the color is the crop and when you hover on any outliers it shows the name of the country and the year. 
   
   * **Hint:** Checkout the [Plotly hover text formatting documentation](https://plotly.com/python/hover-text-and-formatting/)
  
   


````python
# Answer 5
import pandas as pd
import plotly.express as px

farm_data = pd.read_csv('Resources/G20_crop_production_2018-2020.csv')
farm_data
wh_ri_so = farm_data[farm_data.Crop.isin(['Wheat', 'Rice', 'Soybean'])]
wh_ri_so.head()


farm_plot_fig = px.box(wh_ri_so, y='Value(tonnes of HA)', color='Crop', hover_name='Country', hover_data = ['Country','Year'], 
                            color_discrete_map={'Wheat': 'orangered', 'Rice': 'cornflowerblue', 'Soybean': 'navy'})
farm_plot_fig.show()
````

**Question 10: Create a code block that performs the following operations: (15 points)**

1. Import the `earthquakes_database.csv` file into a Pandas dataframe named `earthquakes`

    * **Hint:** There should be 13 data sources.

    * **Note:** You will need to filter the `earthquakes` dataframe for each category type.

````python
#  Answer 10
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
earthquakes = pd.read_csv("Resources/earthquakes_database.csv")
earthquakes.groupby('Source').nunique()
Sources = earthquakes[earthquakes.Source.isin(['AK', 'ATLAS', 'CI','GCMT','ISCGEM','ISCGEMSUP','NC','NN','OFFICIAL','PR','SE','US','UW'])]
px.scatter(
    Sources,
    x='Depth',
    y='Magnitude',
    color='Source',
    color_discrete_map={'AK': 'orangered'},
    hover_name='Source',
)
Source1 = earthquakes[earthquakes.Source.isin(['AK'])]
px.scatter(
    Source1,
    x='Depth',
    y='Magnitude',
    title='AK',
    color='Source',
    color_discrete_map={'AK': 'orangered'},
    hover_name='Source'
)
#plt.scatter(x, y, c ="blue")
 
# To show the plot
#plt.show()
#, 'ATLAS': 'cornflowerblue', 'CI': 'navy','GCMT':'pink','ISCGEM':'black','ISCGEMSUP':'brown','NC':'green','NN':'purple','OFFICIAL':'violet','PR':'white','US':'Lime','UW':'grey'},
````
